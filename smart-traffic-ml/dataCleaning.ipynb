```python
import pandas as pd
train_data = pd.read_csv('train_motion_data.csv')
test_data = pd.read_csv('test_motion_data.csv')

print(train_data.head())
print(test_data.info())
print(train_data['Class'].value_counts())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Seaborn
sns.set(style="whitegrid")

#  (AccX, AccY, AccZ) 
plt.figure(figsize=(12, 6))
plt.plot(train_data['Timestamp'], train_data['AccX'], label='AccX')
plt.plot(train_data['Timestamp'], train_data['AccY'], label='AccY')
plt.plot(train_data['Timestamp'], train_data['AccZ'], label='AccZ')
plt.legend(loc='upper right')
plt.title('Acceleration Data over Time')
plt.xlabel('Timestamp')
plt.ylabel('Acceleration')
plt.xticks(rotation=45)
plt.show()
```
